## Data Cleaning

```python
#Outlier detection
#using IQR
#Chances of getting outliers are only in the 'Age' and 'Salary' columns.
```

```python
#outlier detection in 'Age'
Q1=df['Age'].quantile(0.25)
Q3=df['Age'].quantile(0.75)
IQR=Q3-Q1
lower_limit=Q1-1.5*IQR
upper_limit=Q3+1.5*IQR
```

```python
upper_limit,lower_limit
```

```python
df[(df['Age']>upper_limit) | (df['Age']<lower_limit)]
```

```python
#No outliers in age column but age=0 is an outlier
```

```python
#outliers in salary
Q1=df['Salary'].quantile(0.25)
Q3=df['Salary'].quantile(0.75)
IQR=Q3-Q1
lower_limit=Q1-1.5*IQR
upper_limit=Q3+1.5*IQR
```

```python
upper_limit,lower_limit
```

```python
df[(df['Salary']>upper_limit) | (df['Salary']<lower_limit)]
```

```python
#now treating missing values
df.isnull().sum()
```

```python
#replacing nan with most repeated value i.e TCS
df['Company'].mode(0)
```

```python
df['Company'].fillna(df['Company'].mode()[0],inplace=True)
```

```python
df.isnull().sum()
```

```python
#replacing 0 age with mean (first replace 0 as nan)
df.describe()
```

```python
#replace 0 as 'NaN'
df['Age'].replace(0,np.nan,inplace=True)
```

```python
#mean of age is 30.5 round it
round_mean=round(df['Age'].mean())

df['Age'].fillna(round_mean,inplace=True)
```

```python
df.isnull().sum()
```

```python
#same for salary

round_mean=round(df['Salary'].mean(),0)

df['Salary'].fillna(round_mean,inplace=True)
```

```python
df.isnull().sum()
```

```python
df['Place'].mode(0)
```

```python
df['Place'].fillna(df['Place'].mode()[0],inplace=True)
```

```python
df.isnull().sum()
```

```python
#Now dataset is clean
```
